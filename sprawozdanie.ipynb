**Wielomian**

Dany jest wielomian $w(x)$ w postaci kanonicznej (ogólnej):

$$
w(x) = a_nx^n + a_{n-1}x^{n-1} + \ldots + a_2x^2 + a_1x + a_0
$$

jego wektor współczynników wyraża się wzorem:

$$
\mathbf{a} = (a_0, a_1, ..., a_{n-1}, a_n)
$$

Wnioski do Zadania 2:

Problematyka pierwiastków wielokrotnych:
Numeryczne wyznaczanie pierwiastków wielokrotnych jest obarczone bardzo dużym błędem. Komputer nie jest w stanie odróżnić jednego pierwiastka o krotności $n$ od grupy $n$ różnych pierwiastków leżących bardzo blisko siebie.

Dla pierwiastka o krotności $n$, błąd numeryczny $\epsilon$ (precyzja maszyny) przekłada się na błąd wyniku rzędu $\sqrt[n]{\epsilon}$. Dla $n=8$ oznacza to, że błąd jest wielokrotnie większy niż standardowa dokładność float, co skutkuje ułożeniem się wyników w charakterystyczny okrąg na płaszczyźnie zespolonej zamiast w jednym punkcie.

W celu przeprowadzenia analizy skonstruuj trzy macierze diagonalne $\mathbf{A}_n$ dla $n = \{10,20,30\}$. Współczynniki macierzy powinny być zdefiniowane jako $a_{ii} = 2^i$ dla $i = \{1,2,...,n\}$ (skorzystaj z funkcji [`np.diag()`](https://numpy.org/doc/stable/reference/generated/numpy.diag.html)).

**Punkt 5.**

Porównaj wyniki otrzymane we wszystkich punktach. Opisz z czego wynikają powstałe błędy obliczeń.

````python
# ===== Twoja implementacja tutaj =====
n_values = [10, 20, 30]

errors_A = {'max': [], 'mean': []}
errors_B = {'max': [], 'mean': []}
errors_C = {'max': [], 'mean': []}
errors_F = {'max': [], 'mean': []}

for n in n_values:
    powers = np.arange(1, n + 1)
    eigenvalues_analytical = 2.0 ** powers
    eigenvalues_analytical_sorted = np.sort(eigenvalues_analytical)

    A = np.diag(eigenvalues_analytical)
    vals_A = np.linalg.eigvals(A)
    vals_A_sorted = np.sort(np.real(vals_A))
    diff_A = np.abs(vals_A_sorted - eigenvalues_analytical_sorted)
    errors_A['max'].append(np.max(diff_A))
    errors_A['mean'].append(np.mean(diff_A))

    while True:
        P = np.random.randint(0, 101, (n, n))
        if main.is_nonsingular(P):
            break
            
    P_inv = np.linalg.inv(P)
    B = P @ A @ P_inv
    vals_B = np.linalg.eigvals(B)
    vals_B_sorted = np.sort(np.real(vals_B))
    diff_B = np.abs(vals_B_sorted - eigenvalues_analytical_sorted)
    errors_B['max'].append(np.max(diff_B))
    errors_B['mean'].append(np.mean(diff_B))

    Q, R = np.linalg.qr(P)
    C = Q @ A @ Q.T
    vals_C = np.linalg.eigvals(C)
    vals_C_sorted = np.sort(np.real(vals_C))
    diff_C = np.abs(vals_C_sorted - eigenvalues_analytical_sorted)
    errors_C['max'].append(np.max(diff_C))
    errors_C['mean'].append(np.mean(diff_C))

    try:
        coefs = nppoly.polyfromroots(eigenvalues_analytical)
        F = main.frob_a(coefs)
        
        if F is None:
            raise ValueError
            
        vals_F = np.linalg.eigvals(F)
        vals_F_sorted = np.sort(np.real(vals_F))
        diff_F = np.abs(vals_F_sorted - eigenvalues_analytical_sorted)
        errors_F['max'].append(np.max(diff_F))
        errors_F['mean'].append(np.mean(diff_F))
        
    except Exception:
        errors_F['max'].append(float('nan'))
        errors_F['mean'].append(float('nan'))

print(f"{'Macierz':<25} {'Błąd maksymalny':<20} {'Błąd średni':<20}")
print("-" * 65)

for i, n in enumerate(n_values):
    print(f"A_{n:<3} (diag)             {errors_A['max'][i]:<20.2e} {errors_A['mean'][i]:<20.2e}")
    print(f"B_{n:<3} (P*A*P^-1)         {errors_B['max'][i]:<20.2e} {errors_B['mean'][i]:<20.2e}")
    print(f"C_{n:<3} (Q*A*Q^T)          {errors_C['max'][i]:<20.2e} {errors_C['mean'][i]:<20.2e}")
    print(f"F_{n:<3} (Frobenius)        {errors_F['max'][i]:<20.2e} {errors_F['mean'][i]:<20.2e}")
    print()
````

Wnioski Zadanie 4

Obliczenia wykorzystujące macierze ortogonalne (Q) są znacznie bardziej stabilne numerycznie i generują mniejsze błędy niż transformacje wykorzystujące zwykłe macierze odwracalne (P). Dzieje się tak, ponieważ macierze ortogonalne nie zmieniają długości wektorów, przez co nie "rozdmuchują" błędów zaokrągleń. Metoda wyznaczania wartości własnych poprzez wielomian charakterystyczny (macierz Frobeniusa) jest skrajnie niestabilna. Dla dużych stopni wielomianu współczynniki rosną gwałtownie, prowadząc do utraty precyzji lub przekroczenia zakresu pamięci komputera.

Źródło błędów: Każda operacja numeryczna, a zwłaszcza odwracanie macierzy, wprowadza błędy. Im bardziej skomplikowane przekształcenia wykonujemy na macierzy wejściowej, tym bardziej wynik końcowy odbiega od analitycznego wzorca.
